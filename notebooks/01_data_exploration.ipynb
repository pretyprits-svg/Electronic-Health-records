# Medical NLP Project - Data Exploration

This notebook explores the medical datasets and demonstrates basic text analysis.

Based on: "Natural language processing techniques applied to the electronic health record in clinical research and practice"

```python
# Import required libraries
import sys
sys.path.append('../src')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from utils import load_data, print_sample_data, plot_class_distribution, plot_text_length_distribution

# Set style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

print("Libraries imported successfully!")
```

## 1. Load Sample Data

```python
# Load sample healthcare data
try:
    df_healthcare = pd.read_csv('../data/sample/sample_healthcare.csv')
    print("✓ Healthcare data loaded successfully")
except FileNotFoundError:
    print("✗ Healthcare data not found. Run setup_data.py first.")
    df_healthcare = None

# Load sample medical text data
try:
    df_medical_text = pd.read_csv('../data/sample/sample_medical_text.csv')
    print("✓ Medical text data loaded successfully")
except FileNotFoundError:
    print("✗ Medical text data not found. Run setup_data.py first.")
    df_medical_text = None
```

## 2. Explore Healthcare Dataset

```python
if df_healthcare is not None:
    print_sample_data(df_healthcare)
```

```python
if df_healthcare is not None:
    # Plot admission type distribution
    plot_class_distribution(df_healthcare['Admission Type'].tolist(), 
                           title='Distribution of Admission Types')
```

```python
if df_healthcare is not None:
    # Plot medical condition distribution
    plot_class_distribution(df_healthcare['Medical Condition'].tolist(), 
                           title='Distribution of Medical Conditions')
```

## 3. Explore Medical Text Dataset

```python
if df_medical_text is not None:
    print_sample_data(df_medical_text)
```

```python
if df_medical_text is not None:
    # Plot text length distribution
    plot_text_length_distribution(df_medical_text['text'].tolist(), 
                                 title='Distribution of Medical Text Lengths')
```

```python
if df_medical_text is not None:
    # Plot label distribution
    plot_class_distribution(df_medical_text['label'].tolist(), 
                           title='Distribution of Medical Categories')
```

## 4. Sample Clinical Notes Analysis

```python
if df_healthcare is not None and 'Clinical Notes' in df_healthcare.columns:
    print("Sample Clinical Notes:\n")
    for i, note in enumerate(df_healthcare['Clinical Notes'].head(3), 1):
        print(f"{i}. {note}\n")
```

## 5. Summary Statistics

```python
if df_healthcare is not None:
    print("Healthcare Dataset Summary:")
    print(df_healthcare.describe())
```
