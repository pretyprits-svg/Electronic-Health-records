## 1. Sample Medical Text

Let's start with realistic medical notes:

```python
# Sample medical texts
medical_texts = [
    "Patient presents with severe chest pain and shortness of breath. History of hypertension and diabetes.",
    "68-year-old male with acute myocardial infarction. Patient is experiencing diaphoresis and nausea.",
    "Female patient complaining of persistent headache and dizziness. No history of migraines.",
    "Patient has been diagnosed with Type 2 Diabetes Mellitus. Currently on Metformin 500mg twice daily.",
    "Post-operative patient recovering well. No signs of infection. Vital signs stable."
]

print("📝 Sample Medical Texts:")
print("=" * 80)
for i, text in enumerate(medical_texts, 1):
    print(f"\n{i}. {text}")
print("\n" + "=" * 80)
```

## 2. Initialize Preprocessor

Create the Medical Text Preprocessor with default settings:

```python
# Initialize preprocessor
preprocessor = MedicalTextPreprocessor()

print("✅ MedicalTextPreprocessor initialized!")
print("\nPreprocessor Configuration:")
print(f"  - Lowercase: True")
print(f"  - Remove stopwords: True")
print(f"  - Lemmatization: True")
print(f"  - Remove punctuation: True")
```

## 3. Step-by-Step Preprocessing

Let's process one text and see each step:

```python
# Take first text as example
sample_text = medical_texts[0]

print("📄 ORIGINAL TEXT:")
print("=" * 80)
print(sample_text)
print("=" * 80)

# Process it
processed = preprocessor.preprocess(sample_text)

print("\n✨ PROCESSED TEXT:")
print("=" * 80)
print(processed)
print("=" * 80)

print("\n📊 STATISTICS:")
print(f"  Original length: {len(sample_text)} characters")
print(f"  Processed length: {len(processed)} characters")
print(f"  Original words: {len(sample_text.split())}")
print(f"  Processed words: {len(processed.split())}")
print(f"  Reduction: {100 - (len(processed.split())/len(sample_text.split())*100):.1f}%")
```

## 4. Process All Texts

Now let's process all medical texts:

```python
# Process all texts
processed_texts = [preprocessor.preprocess(text) for text in medical_texts]

# Create comparison DataFrame
df_comparison = pd.DataFrame({
    'Original': medical_texts,
    'Processed': processed_texts,
    'Original_Length': [len(t) for t in medical_texts],
    'Processed_Length': [len(t) for t in processed_texts],
    'Word_Count_Before': [len(t.split()) for t in medical_texts],
    'Word_Count_After': [len(t.split()) for t in processed_texts]
})

print("📊 Preprocessing Results:")
print(df_comparison[['Word_Count_Before', 'Word_Count_After']].describe())
```
