```python
# Import all required libraries
import sys
sys.path.append('../src')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from wordcloud import WordCloud

from preprocessing import MedicalTextPreprocessor, preprocess_text
from anonymization import MedicalTextAnonymizer, anonymize_text
from bow_model import BagOfWordsModel
from tfidf_model import TFIDFModel
from utils import create_wordcloud, compare_model_performance

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (14, 8)

print("✅ All libraries imported successfully!")
```

## Step 1: Load Data

```python
# Create sample medical dataset
medical_data = {
    'patient_id': range(1, 21),
    'text': [
        "Patient John Smith presents with severe chest pain and dyspnea. DOB: 03/15/1975.",
        "Mary Johnson diagnosed with Type 2 Diabetes. Phone: (555) 123-4567.",
        "Patient experiencing acute myocardial infarction. Elevated troponin levels.",
        "Robert Williams has persistent headache and visual disturbances.",
        "Patient with fractured left femur. Requires immediate surgical intervention.",
        "Sarah Davis complaining of severe abdominal pain and nausea.",
        "Patient has history of hypertension and hyperlipidemia. BP: 160/95.",
        "Michael Brown diagnosed with pneumonia. Chest X-ray shows infiltrates.",
        "Patient presents with symptoms of urinary tract infection.",
        "Jennifer Wilson has chronic kidney disease stage 3. Creatinine: 2.1.",
        "Patient experiencing severe allergic reaction. Administered epinephrine.",
        "David Martinez with acute appendicitis. Scheduled for appendectomy.",
        "Patient has deep vein thrombosis in right leg. Started on anticoagulation.",
        "Lisa Anderson diagnosed with rheumatoid arthritis. Joint pain and swelling.",
        "Patient with acute asthma exacerbation. Wheezing and shortness of breath.",
        "James Taylor has gastroesophageal reflux disease. Prescribed omeprazole.",
        "Patient experiencing symptoms of depression and anxiety.",
        "Patricia Thomas diagnosed with osteoporosis. T-score: -2.8.",
        "Patient with acute stroke. CT scan shows ischemic changes.",
        "Christopher Lee has chronic obstructive pulmonary disease. FEV1: 45%."
    ],
    'category': [
        'Cardiac', 'Endocrine', 'Cardiac', 'Neurological', 'Orthopedic',
        'Gastrointestinal', 'Cardiac', 'Respiratory', 'Urological', 'Renal',
        'Immunological', 'Gastrointestinal', 'Vascular', 'Rheumatological', 'Respiratory',
        'Gastrointestinal', 'Psychiatric', 'Orthopedic', 'Neurological', 'Respiratory'
    ]
}

df = pd.DataFrame(medical_data)

print(f"📊 Dataset loaded: {len(df)} patient records")
print(f"\nCategories: {df['category'].nunique()} unique")
print(df['category'].value_counts())
df.head()
```

## Step 2: Preprocessing

```python
print("🔄 Step 2: Preprocessing medical text...")

preprocessor = MedicalTextPreprocessor()
df['preprocessed'] = df['text'].apply(lambda x: preprocessor.preprocess(x))

print("✅ Preprocessing complete!")
print("\nExample:")
print(f"Original: {df['text'].iloc[0]}")
print(f"Preprocessed: {df['preprocessed'].iloc[0]}")
```

## Step 3: Anonymization

```python
print("🔄 Step 3: Anonymizing PHI/PII...")

anonymizer = MedicalTextAnonymizer()
df['anonymized'] = df['text'].apply(lambda x: anonymizer.anonymize(x))

print("✅ Anonymization complete!")
print("\nExample:")
print(f"Original: {df['text'].iloc[0]}")
print(f"Anonymized: {df['anonymized'].iloc[0]}")
```

## Step 4: Train-Test Split

```python
X = df['preprocessed'].tolist()
y = df['category'].tolist()

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print(f"📊 Training samples: {len(X_train)}")
print(f"📊 Testing samples: {len(X_test)}")
```

## Step 5: Model Training & Evaluation

```python
print("🔄 Step 5: Training models...\n")

# Bag of Words
print("Training Bag of Words...")
bow_model = BagOfWordsModel(max_features=30)
X_train_bow = bow_model.fit_transform(X_train)
X_test_bow = bow_model.transform(X_test)
bow_model.train_classifier(X_train_bow, y_train)
bow_results = bow_model.evaluate(X_test_bow, y_test)

print(f"✅ BoW Accuracy: {bow_results['accuracy']:.2%}\n")

# TF-IDF
print("Training TF-IDF...")
tfidf_model = TFIDFModel(max_features=30)
X_train_tfidf = tfidf_model.fit_transform(X_train)
X_test_tfidf = tfidf_model.transform(X_test)
tfidf_model.train_classifier(X_train_tfidf, y_train)
tfidf_results = tfidf_model.evaluate(X_test_tfidf, y_test)

print(f"✅ TF-IDF Accuracy: {tfidf_results['accuracy']:.2%}")
```

## Step 6: Results Summary

```python
print("\n" + "="*70)
print("PIPELINE COMPLETE - RESULTS SUMMARY")
print("="*70)
print(f"\n📊 Total Records Processed: {len(df)}")
print(f"📊 Categories: {df['category'].nunique()}")
print(f"\n🤖 Model Performance:")
print(f"  - Bag of Words: {bow_results['accuracy']:.2%}")
print(f"  - TF-IDF: {tfidf_results['accuracy']:.2%}")
print(f"\n✅ All steps completed successfully!")
```
